```python
# Ê≥®ÊÑèÔºÅÊ≠§Â§ÑÁöÑRGBÈ¢úËâ≤ÊòØÊ†πÊçÆEmojiÁ¨¶Âè∑Âú®"Windows 10 May 2019 Update"ÁâàÊú¨‰∏äÁöÑÈïøÁõ∏ËÄåÂÆöÁöÑ„ÄÇÂØπ‰∫éios„ÄÅmac„ÄÅÂÆâÂçìÁ≠âÂπ≥Âè∞ÔºåÈ¢úËâ≤ÂèØËÉΩ‰ºö‰∏ç‰∏ÄÊ†∑„ÄÇ
emoji2RGB = {"‚ù§Ô∏è":[232,18,36],"üòÉ":[255,200,61],"ü§¢":[140,189,24],
             "üßä":[179,219,242],"üìò":[58,150,221],"üçá":[92,46,145],
             "üí©":[142,86,46],"üê∑":[238,152,137],"üêò":[204,204,204],
             "üêá":[255,255,255],u"üï∂Ô∏è":[32,32,32],"üê™":[255,200,61],
             "üêå":[247,177,137],"üêñ":[238,152,137],"üêØ":[247,99,12]}
```

```python
# ËæÖÂä©ÂáΩÊï∞ÔºåÊâæÂá∫‰∏éÊåáÂÆöÈ¢úËâ≤ÊúÄËøëÁöÑEmoji
def findNearestEmoji(color,emojiList):
    bestDistance = 255*3
    bestEmoji = ""
    for emoji in emojiList:
        emojiColor = emojiList[emoji]
        distance = sum(abs(color - emojiColor))
        if distance<bestDistance:
            bestEmoji = emoji
            bestDistance = distance
    return bestEmoji
```

```python
# ÂõæÁâáËΩ¨Êç¢‰∏∫Emoji
emojiAnswer = ""
for y in range(OUTPUT_H):
    for x in range(OUTPUT_W):
        imgColor = mabaoguoAfterResize[y,x]
        emojiAnswer += findNearestEmoji(imgColor,emoji2RGB)
    emojiAnswer += "\n"
```

```python
print(emojiAnswer)
```

```python
# ÂõæÂÉèËß£ÂØÜÁÆóÊ≥ï„ÄÇËÄÉËôëÂà∞‰∫ÜÊúâ‰∫õUnicodeÂ≠òÂÇ®Â§ßÂ∞èÊòØÂÖ∂‰ªñUnicode‰∏§ÂÄçÁöÑÂèØËÉΩ„ÄÇÂõ†Ê≠§ÁÆóÊ≥ïËæÉ‰∏∫Â§çÊùÇ
def emojiImgDecode(text,emoji2RGB):
    imgDecode = np.zeros((OUTPUT_H,OUTPUT_H,3))
    currentIndex = 0
    for emojiID in range(OUTPUT_H*(OUTPUT_W+1)):
        emojiChar = text[currentIndex]
        currentIndex+=1
        if emojiChar=="\n":
            continue
        if emojiChar not in emoji2RGB:
            emojiChar = text[currentIndex-1:currentIndex+1]
            currentIndex+=1
        y = emojiID//(OUTPUT_W+1)
        x = emojiID%(OUTPUT_W+1)
        imgDecode[y,x,:] = emoji2RGB[emojiChar]
        imgDecode = imgDecode.astype('uint8')
    return imgDecode
```

```python
imgDecode = emojiImgDecode(emojiAnswer,emoji2RGB)
```

```python
plt.figure(figsize = (12,8))
plt.subplot(121)
plt.title(u"Âä†ÂØÜÂâç",fontsize=16)
plt.imshow(mabaoguoAfterResize)
plt.subplot(122)
plt.title(u"Âä†ÂØÜÂÜçËß£ÂØÜÂêé",fontsize=16)
plt.imshow(imgDecode)
```

```python
def caesarEncode(pinyin,k):
    afterEncode = ""
    for character in pinyin:
        afterEncode+=chr((ord(character)-ord("a")+k)%26+ord("a"))
    return afterEncode
```

```python
def emojiCaesarEncode(text):
    returnEmoji = ""
    for character in text:
        ID = ord(character)-ord("a")
        offset = int.from_bytes(b'\xf0\x9f\x90\x80',byteorder="big")
        newID = offset+ID
        newEmoji = newID.to_bytes(length=4,byteorder="big")
        returnEmoji+=newEmoji.decode()
    return returnEmoji
```

### Level 4 Êë©Â∞îÊñØÂä†ÂØÜ

```python
morseChar2Code = {"a":".-","b":"-...","c":"-.-.","d":"-..",
                  "e":".","f":"..-.","g":"--.","h":"....",
                  "i":"..","j":".---","k":"-.-","l":".-..",
                  "m":"--","n":"-.","o":"---","p":".--.",
                  "q":"--.-","r":".-.","s":'...',"t":"-",
                  "u":"..-","v":"...-","w":".--","x":"-..-",
                  "y":"-.--","z":"--..","1":".----","2":"..---",
                  "3":"...--","4":"....-","5":".....","6":"-....",
                  "7":"--...","8":"---..","9":"----.","0":"-----",
                 ".":".-.-.-",",":"--..--","?":"..--..","!":"..--."}
morseCode2Char = {}
for key in morseChar2Code:
    morseCode2Char[morseChar2Code[key]] = key
```

```python
def morseEncode(text):
    returnText = ""
    for character in text:
        returnText+=morseChar2Code[character]+" "
    return returnText
```

```python
print("Âä†ÂØÜÂâçÔºö",pinyin)
print("Âä†ÂØÜÂêéÔºö",morseEncode(pinyin))
```

```python
morseEncode("wan8dianlaodifang")
```

```python
def morseDecode(morseCode):
    return "".join([morseCode2Char[code] for code in morseCode.split(" ") \
                    if code in morseCode2Char])
```

```python
morseDecode(morseEncode(pinyin))
```

```python
import random
```

```python
def morse2Emoji(morseCode):
    returnEmoji = b""
    humanID = skinID = 0
    for morseChar in morseCode:
        humanID = random.randint(0,14)
        if morseChar ==".":
            skinID = 1
        elif morseChar =="-":
            skinID = 5
        else:
            skinID = 3
        returnEmoji += HumanCharList[humanID] + skinColorList[skinID]
    return returnEmoji.decode("utf-32")
```
